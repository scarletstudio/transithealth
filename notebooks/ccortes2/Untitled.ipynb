```python
import pandas as pd
import requests

dataset_json_url = "https://data.cityofchicago.org/resource/jyb9-n7fm.json"
query = """
SELECT
    route,
    date,
    daytype,
    rides
WHERE
    date >= "2020-01-01" AND date <= "2021-03-31"
LIMIT 800000
"""
r = requests.get(dataset_json_url, params={"$query": query})
df=pd.DataFrame(r.json())
```
